*Import Libary*

```python
from google_play_scraper import reviews, reviews_all, Sort
import pandas as pd
```

*Melakukan scraping ulasan aplikasi shopee*

```python
scrapreview = reviews (
    'com.shopee.id',       # Shopee ID
    lang='id',             # Bahasa Indonesia
    country='id',          # Negara Indonesia
    sort=Sort.NEWEST,      # Review terbaru
    count=12000           # Jumlah ulasan yang ingin diambil
)
```

*Menyimpan hasilnya menjadi file csv*

```python
# scrapreview adalah tuple (data, continuation_token)
df = pd.DataFrame(scrapreview[0])                       # ambil index 0 untuk datanya
df = df[['userName', 'content', 'score', 'at']]
df.to_csv('ulasan_aplikasi_shopee.csv', index=False)
print(f"Total data tersimpan: {len(df)}")
```
